```python
import numpy as np
from astropy.time import Time
import formatting
import telescope
import pandas as pd

startObs = Time('2024-02-11T07:00:00', format='isot')
endObs = Time('2024-02-11T14:00:00', format='isot')
time = np.linspace(startObs.jd,endObs.jd,100)
tel = telescope.create_tel('HALE',3)

# Same format of target list you supply to the full model
all_stars = formatting.theTTP('../tutorial/TKStargets.csv')

texp_df = pd.DataFrame()
texp_df['jd'] = time

for i in range(len(all_stars)):
    test_star = all_stars[i]

    airmasses = np.array(tel.observer.altaz(Time(time,format='jd'), test_star.target).secz)
    print(test_star.name,airmasses)

    # Insert ETC Code here to replace this section
    texp = test_star.exptime*airmasses**3
    #texp = test_star.exptime*np.ones(len(airmasses))
    # The .exptime attribute is in minutes. The exposure time dataframe will also be assumed to be in minutes


    below_horizon = np.where(airmasses < 0)[0]
    texp[below_horizon] = np.nan

    texp_df[test_star.name] = texp

texp_df.to_csv('../tutorial/texp.csv')
```
